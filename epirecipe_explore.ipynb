```python
integer_columns = df.columns[6:] 

temp = {column: int for column in integer_columns}
df = df.astype(temp)
df = df[(df['rating'] >= 0) & (df['rating'] <= 5)] 
df = df[df['calories'] < 10000]  
df = df[df['protein'] < 5000]    
df = df[df['fat'] < 5000]        
df = df[df['sodium'] < 5000]   
df.dropna(inplace=True)
tags = df.columns[6:] 
df['tags_count'] = df[tags].sum(axis=1) 
df.drop(columns=['title'], axis=1, inplace=True)
print(df.head())
```

```python
def print_unique_and_nunique(df1):
    for column in df.columns:
        unique_values = df[column].unique()  # Get unique values in the column
        unique_count = df[column].nunique()  # Get count of unique values in the column
        print(f"Column '{column}':")
        print(f"  Unique values: {unique_values}")
        print(f"  Number of unique values: {unique_count}")
        print('-' * 50)
print_unique_and_nunique(df1)
```

```python
def remove_outliers_iqr(df1):
    for column in df.select_dtypes(include=[np.number]).columns:
        Q1 = df1[column].quantile(0.25)
        Q3 = df1[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df1 = df1[(df1[column] >= lower_bound) & (df1[column] <= upper_bound)]
    return df1

df_no_outliers_iqr = remove_outliers_iqr(df1)
print("DataFrame after removing outliers using IQR:")
print(df_no_outliers_iqr)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df1' contains the data
df = pd.DataFrame(df1)

# 1. Rating Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=10, kde=True)
plt.title('Rating Distribution of Recipes')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid()
plt.show()

# 2. Caloric Trends: Average Calories by Rating
caloric_trends = df.groupby('rating')['calories'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=caloric_trends, x='rating', y='calories', palette='pastel')
plt.title('Average Caloric Content by Recipe Rating')
plt.xlabel('Rating')
plt.ylabel('Average Calories')
plt.grid()
plt.show()

# 3. Protein Content Analysis: Average Protein by Rating
protein_trends = df.groupby('rating')['protein'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=protein_trends, x='rating', y='protein', palette='Set2')
plt.title('Average Protein Content by Recipe Rating')
plt.xlabel('Rating')
plt.ylabel('Average Protein (g)')
plt.grid()
plt.show()

# 4. Fat and Sodium Correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='fat', y='sodium', hue='rating', size='calories', sizes=(20, 200), palette='coolwarm')
plt.title('Fat vs Sodium in Recipes')
plt.xlabel('Fat (g)')
plt.ylabel('Sodium (mg)')
plt.legend(title='Rating', loc='upper left')
plt.grid()
plt.show()
```
